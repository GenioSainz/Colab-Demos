
# Binary Cross-Entropy
$$
C_{BCE}  = -\frac{1}{n} \sum_x \sum_j y_j \ln(a^L_j)+ (1-y_j) \ln(1-a^L_j)
$$

# Logistic regression
$$
sigmoid(x)  = \frac{1}{1+e^{-(k·x+b)}} = probability
$$

$$
sigmoid(k·x+b) = probability
$$

```python

# @title Entropy
%matplotlib inline

N  = 75
R1 = 0.25
C1 = 0.25
R2 = 0.45
C2 = 0.45

theta1 = np.random.rand(N,2)*2*np.pi
r1     = np.random.rand(N,2)*R1

theta2 = np.random.rand(N,2)*2*np.pi
r2     = np.random.rand(N,2)*R2

x1r,y1r = C1     + r1[:,0]*np.cos(theta1[:,0]), C1     + r1[:,0]*np.sin(theta1[:,0])
x1g,y1g = (1-C1) + r1[:,1]*np.cos(theta1[:,1]), (1-C1) + r1[:,1]*np.sin(theta1[:,1])

x2r,y2r = C2     + r2[:,0]*np.cos(theta2[:,0]), C2     + r2[:,0]*np.sin(theta2[:,0])
x2g,y2g = (1-C2) + r2[:,1]*np.cos(theta2[:,1]), (1-C2) + r2[:,1]*np.sin(theta2[:,1])

sizeM     = 50
fig2, ax2 = plt.subplots(1,2, constrained_layout=True,figsize=(800*px2inch , 400*px2inch))

ax2[0].scatter(x1r,y1r,sizeM,'r',alpha=0.5)
ax2[0].scatter(x1g,y1g,sizeM,'g',alpha=0.5)
plot_sets(ax2[0],ti='Low Entropy',xli=[0,1],yli=[0,1])

ax2[1].scatter(x2r,y2r,sizeM,'r',alpha=0.5)
ax2[1].scatter(x2g,y2g,sizeM,'g',alpha=0.5)
plot_sets(ax2[1],ti='High Entropy',xli=[0,1],yli=[0,1])
```

```python

# @title Main code
from   sklearn.linear_model import LogisticRegression
from   sklearn.metrics      import log_loss
import numpy as np

def sigmoid(z):

    return 1/(1+np.exp(-z))

def cross_entropy(y,a):

    return -( y*np.log(a) + (1-y)*np.log(1-a) )

def cross_entropy_sum(y,a):

    return np.sum( -( y*np.log(a) + (1-y)*np.log(1-a)) ) / a.size

# original
# y = np.array([0   ,  0   , 1  ,  0  , 1  , 0  , 1  , 1  , 1  , 1  ])
# a = np.array([-1.8, -0.8, -0.5, -0.1, 0.5, 0.6, 0.9, 1.1, 1.5, 1.9])

y = np.array([0   ,0   ,  0  , 1   ,  0  , 1  ,0  , 1  , 1  , 1  , 1  ,1])
a = np.array([-2.5,-1.8, -0.8, -0.5, -0.1, 0.5,0.6, 0.9, 1.1, 1.5, 1.9,2.5])

logr = LogisticRegression()
logr.fit(a.reshape(-1, 1), y)

y_pred = logr.predict_proba(a.reshape(-1, 1))[:, 1].ravel()

x_regression = np.linspace(a[0],a[-1],100)
y_regression = sigmoid(x_regression * logr.coef_ + logr.intercept_).ravel()

loss1 = log_loss(y, y_pred)
loss2 = cross_entropy_sum(y,y_pred)

print('a                        = {}'.format(a))
print('y                        = {}'.format(y))
print('Logistic Regression p(y) = {}'.format(np.round(y_pred, 2)))
print('Cross Entropy sklearn    = {:.4f}'.format(loss1))
print('Cross Entropy compute    = {:.4f}'.format(loss2))


%matplotlib inline
fig0, ax0 = plt.subplots(3,1, constrained_layout=True,figsize=(1300*px2inch , 800*px2inch))

indx_pos = y==1
indx_neg = y==0
ax0[0].plot([np.min(a),np.max(a)],[0,0],'r--');
ax0[0].plot([np.min(a),np.max(a)],[1,1],'g--');
ax0[0].plot([0,0],[0,1],'k');

ax0[0].plot(x_regression,y_regression ,'b--',label='LogisticRegression');
ax0[0].plot(a,y_pred                  ,'bo' ,label='Samples');
ax0[0].plot(a[indx_pos],y[indx_pos]   ,'go' ,label='R Class');
ax0[0].plot(a[indx_neg],y[indx_neg]   ,'ro' ,label='G Class');

for (ai,yi,txt_dy) in zip(a,y_pred,y):

    dy = -0.1
    if txt_dy == 0: 
        dy=-dy
    ax0[0].text(ai,yi+dy,'{:.3f}'.format(yi),ha='center',va='center')

plot_sets(ax0[0],gr=True,xla='x',yla='p(x)',leg=dict(ncol=4,loc="upper center",bbox_to_anchor=(0.5, 1.25)))

classification = y*y_pred + (1-y)*(1-y_pred)
bar_colors     = ["green" if n==1  else "red" for n in y]

bar1 = ax0[1].bar(np.arange(a.size),classification,color=bar_colors);
ax0[1].bar_label(bar1,fmt='%.3f');
ax0[1].set_title('Classification Probability')
ax0[1].set_ylim(0,1.2)
ax0[1].set_xticks(np.arange(a.size),a)
ax0[1].set_ylabel('p(x)')
ax0[1].grid(False)


b_c_e      = cross_entropy(y,y_pred)
bar_colors = ["green" if n==1  else "red" for n in y]

bar2 = ax0[2].bar(np.arange(a.size),b_c_e,color=bar_colors);
ax0[2].bar_label(bar2,fmt='%.3f');
ax0[2].set_title('Binary Cross-Entropy Mean:{:.3f}'.format(loss2))
ax0[2].set_ylim(0,np.max(b_c_e)+0.2)
ax0[2].set_xticks(np.arange(a.size),a)
ax0[2].set_ylabel('Cost')
ax0[2].grid(False)
```

```python


def subplot_entropy(maxA,minA,N):

    y = np.hstack((np.zeros(N),np.ones(N)))
    a = np.hstack((-1*np.linspace(maxA,minA,N),np.linspace(minA,maxA,N)))

    logr = LogisticRegression()
    logr.fit(a.reshape(-1, 1), y)

    y_pred = logr.predict_proba(a.reshape(-1, 1))[:, 1].ravel()

    x_regression = np.linspace(a[0],a[-1],100)
    y_regression = sigmoid(x_regression * logr.coef_ + logr.intercept_).ravel()

    loss1 = log_loss(y, y_pred)
    loss2 = cross_entropy_sum(y,y_pred)

    classification = y*y_pred + (1-y)*(1-y_pred)
    b_c_e          = cross_entropy(y,y_pred)

    print('a                        = {}'.format(a))
    print('y                        = {}'.format(y))
    print('Logistic Regression p(y) = {}'.format(np.round(y_pred, 2)))
    print('Cross Entropy compute    = {:.4f}'.format(loss2))
    print('\n')

    return y, a, y_pred, x_regression, y_regression ,loss1, loss2, classification, b_c_e


%matplotlib inline
fig1, ax1 = plt.subplots(3,2, constrained_layout=True,figsize=(1600*px2inch , 800*px2inch))

# SUBPLOT COL 1

N    = 5
maxA = 2.0
minA = 1.5
y, a, y_pred, x_regression, y_regression, loss1, loss2, classification, b_c_e  = subplot_entropy(maxA,minA,N)

bar_colors = ["green" if n==1  else "red" for n in y]

indx_pos = y==1
indx_neg = y==0
ax1[0][0].plot([np.min(a),np.max(a)],[0,0],'r--');
ax1[0][0].plot([np.min(a),np.max(a)],[1,1],'g--');
ax1[0][0].plot([0,0],[0,1],'k');
ax1[0][0].plot(x_regression,y_regression ,'b--',label='Logistic Regression');
ax1[0][0].plot(a,y_pred                  ,'bo' ,label='Samples');
ax1[0][0].plot(a[indx_pos],y[indx_pos]   ,'go' ,label='G Class');
ax1[0][0].plot(a[indx_neg],y[indx_neg]   ,'ro' ,label='R Class');
ax1[0][0].set_xlabel('x')
ax1[0][0].set_ylabel('p(x)')
ax1[0][0].legend(ncol=4,loc="upper center",bbox_to_anchor=(0.5, 1.25));

bar1 = ax1[1][0].bar(np.arange(a.size),classification,color=bar_colors);
ax1[1][0].bar_label(bar1,fmt='%.3f');
ax1[1][0].set_title('Classification Probability')
ax1[1][0].set_ylim(0,1.2)
ax1[1][0].set_xticks(np.arange(a.size),a)
ax1[1][0].set_ylabel('p(x)')
ax1[1][0].grid(False)

bar2 = ax1[2][0].bar(np.arange(a.size),b_c_e,color=bar_colors);
ax1[2][0].bar_label(bar2,fmt='%.3f');
ax1[2][0].set_title('Binary Cross-Entropy Mean:{:.3f}'.format(loss2))
ax1[2][0].set_ylim(0,0.5)
ax1[2][0].set_xticks(np.arange(a.size),a)
ax1[2][0].set_ylabel('Cost')
ax1[2][0].grid(False)


# SUBPLOT COL 2

N    = 5
maxA = 2.0
minA = 0.5
y, a, y_pred, x_regression, y_regression, loss1, loss2, classification, b_c_e  = subplot_entropy(maxA,minA,N)

bar_colors = ["green" if n==1  else "red" for n in y]

indx_pos = y==1
indx_neg = y==0
ax1[0][1].plot([np.min(a),np.max(a)],[0,0],'r--');
ax1[0][1].plot([np.min(a),np.max(a)],[1,1],'g--');
ax1[0][1].plot([0,0],[0,1],'k');
ax1[0][1].plot(x_regression,y_regression ,'b--',label='Logistic Regression');
ax1[0][1].plot(a,y_pred                  ,'bo' ,label='Samples');
ax1[0][1].plot(a[indx_pos],y[indx_pos]   ,'go' ,label='G Class');
ax1[0][1].plot(a[indx_neg],y[indx_neg]   ,'ro' ,label='R Class');
ax1[0][1].set_xlabel('x')
ax1[0][1].set_ylabel('p(x)')
ax1[0][1].legend(ncol=4,loc="upper center",bbox_to_anchor=(0.5, 1.25));

bar3 = ax1[1][1].bar(np.arange(a.size),classification,color=bar_colors);
ax1[1][1].bar_label(bar3,fmt='%.3f');
ax1[1][1].set_title('Classification Probability')
ax1[1][1].set_ylim(0,1.2)
ax1[1][1].set_xticks(np.arange(a.size),a)
ax1[1][1].set_ylabel('p(x)')
ax1[1][1].grid(False)

bar4 = ax1[2][1].bar(np.arange(a.size),b_c_e,color=bar_colors);
ax1[2][1].bar_label(bar4,fmt='%.3f');
ax1[2][1].set_title('Binary Cross-Entropy Mean:{:.3f}'.format(loss2))
ax1[2][1].set_ylim(0,0.5)
ax1[2][1].set_xticks(np.arange(a.size),a)
ax1[2][1].set_ylabel('Cost')
ax1[2][1].grid(False)
```

```python
def sigmoid(z):          return 1/(1+np.exp(-z))
def sigmoid_diff(z):     return sigmoid(z)*(1-sigmoid(z))

def quadratic_cost(a,y): return  ( y-a )**2
def cross_entropy(a,y):  return -( y*np.log(a) + (1-y)*np.log(1-a) )

def delta_mse(z,a,y): return (a-y)*sigmoid_diff(z)
def delta_bce(z,a,y): return (a-y)

fig3, ax3 = plt.subplots(1,2, constrained_layout=True,figsize=(1000*px2inch , 450*px2inch))

z     = np.linspace(-4,4)
a     = sigmoid(z)

y     = 0
mse   = quadratic_cost(sigmoid(z),y)
bce   = cross_entropy( sigmoid(z),y)
d_mse = delta_mse(z,a,y)
d_bce = delta_bce(z,a,y)
ax3[y].plot(z,sigmoid(z),label='$a=\sigma(z)$');ax3[y].plot(z,mse,label='$MSE _ cost$');ax3[y].plot(z,bce,label='$BCE _ cost$')
ax3[y].plot(z,d_mse,label='$\delta_{MSE}$'); ax3[y].plot(z,d_bce,label='$\delta_{BCE}$');
ax3[y].set_title(f'Desire y = {y}');ax3[y].set_xlabel('z');ax3[y].legend();

y     = 1
mse   = quadratic_cost(sigmoid(z),y)
bce   = cross_entropy( sigmoid(z),y)
d_mse = delta_mse(z,a,y)
d_bce = delta_bce(z,a,y)
ax3[y].plot(z,sigmoid(z),label='$a=\sigma(z)$');ax3[y].plot(z,mse,label='$MSE _ cost$');ax3[y].plot(z,bce,label='$BCE _ cost$')
ax3[y].plot(z,d_mse,label="$\delta_{MSE}$"); ax3[y].plot(z,d_bce,label="$\delta_{BCE}$");
ax3[y].set_title(f'Desire y = {y}');ax3[y].set_xlabel('z');ax3[y].legend();
```
