```python
# @title  Screens Ratios

import matplotlib.pyplot as plt
from   matplotlib.patches import Rectangle
import numpy as np

%matplotlib widget
px2inch = 1/plt.rcParams['figure.dpi']
```

```python

def funt_d(funt,t=0):

    h = 1e-9
    return (funt(t+h)-funt(t))/h
```

```python
size_plot = 500
fig, ax4  = plt.subplots(1,1,constrained_layout=True,figsize=(2*size_plot*px2inch,size_plot*px2inch))

def funt(t=0):
    
    x = 5*np.cos(t)
    y = 2*np.sin(t)

    return np.vstack((x,y))

k   = 2
tc  = np.linspace(0,k*np.pi,180*10)
T   = np.linspace(0,k*np.pi,180)
dt  = np.diff(T)[0]

for t in T:

    ft  = funt(t)
    ftd = funt_d(funt,t=t)*dt
    s   = ft + ftd

    lw  = 0.25
    ax4.plot( [ft[0],s[0]] ,[ft[1],s[1] ],'r',lw=lw); # s
    ax4.plot( [ft[0],s[0]] ,[ft[1],ft[1]],'g',lw=lw); # dx
    ax4.plot( [s[0],s[0] ] ,[ft[1],s[1] ],'b',lw=lw); # dy

ax4.set_aspect(1)
```

```python
def funt_d(funt,t=0):

    h = 1e-6
    return (funt(t+h)-funt(t))/h

def funt(t):
    
    x = np.cos(t)
    y = np.sin(t)

    return np.vstack((x,y))
```

```python
nt      = np.arange(50,10050,50)
arc_len = []
for n in nt :
    t   = np.linspace(0,np.pi,n)
    dt  = np.diff(t)[0]
    ft  = funt(t)
    ftd = funt_d(funt,t=t)

    arc_len_ti = np.sum( np.sqrt(np.sum(ftd**2,axis=0))  )*dt
    arc_len.append( arc_len_ti )
```
