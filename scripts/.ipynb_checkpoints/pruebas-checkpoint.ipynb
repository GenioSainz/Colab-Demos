```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import matplotlib.pyplot as plt
import numpy as np
```

```python

t     = np.linspace(0, 5, 500)
phase = 2

def update_plot(A,f,fi):

    y = A*np.sin(2*np.pi*f*t + fi)

    plt.figure(figsize=(8, 6))
    plt.title(f'Sine Wave \nAmp: {A:0.3f}\nFre: {f:0.3f}\nPha: {fi:0.3f}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.xlim([0,5])
    plt.ylim([-2,2])
    
    plt.plot(t,y)
    plt.show()

# Create an interactive form
interact(update_plot,
                     A =widgets.FloatSlider(value=1.0, min=0.25,   max=2.0,   step=0.25, description='Amplitude:'),
                     f =widgets.FloatSlider(value=1.0, min=1.00,   max=5.0,   step=0.25, description='Frequency:'),
                     fi=widgets.FloatSlider(value=0.0, min=-phase, max=phase, step=0.20, description='Phase:',continuous_update=False),
                );
```

```python
import ipywidgets as widgets
from ipywidgets import GridspecLayout
from IPython.display import display

# Create sliders
slider1 = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.1, description='Slider 1:')
slider2 = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.1, description='Slider 2:')
slider3 = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.1, description='Slider 3:')
slider4 = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.1, description='Slider 4:')

# Arrange sliders in a grid layout
grid = GridspecLayout(2, 2)
grid[0, 0] = slider1
grid[0, 1] = slider2
grid[1, 0] = slider3
grid[1, 1] = slider4

# Define function to be called when sliders change
def on_slider_change(change):
        
        fig = plt.figure(1)
        ax = fig.add_axes([0, 0, 1, 1])

        # fig, ax = plt.subplots(1,1,figsize=(6,4))
        # print(f'{change["owner"].description} value: {change["new"]}')
      
    
        A = grid[0,0].value
        y = A* np.sin(2*np.pi*1*t)
    
        ax.plot(t, y)

        fig.canvas.draw()
        print(A)
        plt.show()

# Link function to slider change event
slider1.observe(on_slider_change, names='value')
slider2.observe(on_slider_change, names='value')
slider3.observe(on_slider_change, names='value')
slider4.observe(on_slider_change, names='value')

# Display the grid of sliders
display(grid)
```
