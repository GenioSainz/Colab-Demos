```python
# @title  Screens Ratios

import matplotlib.pyplot as plt
from   matplotlib.patches import Rectangle
import numpy as np

%matplotlib inline
def plot_rectangle(data,ax,display_horizontal=True):

    r,n,d = data['ratio'],data['name'],data['size']*2.54
    h = np.sqrt(d**2/(1+r**2))
    w = h*r
    if display_horizontal:
       t = f'{n:7s}  W:{w:0.2f}  H:{h:0.2f}  W*H:{w*h:0.2f}'
       print(t)
       ax[0].add_patch( Rectangle((0,0),       w, h, fill=False, label=t, color=data['color']) )
       ax[1].add_patch( Rectangle((-w/2,-h/2), w, h, fill=False, label=t, color=data['color']) )
    else:
       t = f'{n:7s}  W:{h:0.2f}  H:{w:0.2f}  W*H:{w*h:0.2f}'
       print(t)
       ax[0].add_patch( Rectangle((0,0),       h, w, fill=False, label=t, color=data['color']) )
       ax[1].add_patch( Rectangle((-h/2,-w/2), h, w, fill=False, label=t, color=data['color']) )

    ax[0].plot(np.nan,np.nan),ax[0].set_aspect(1)
    ax[1].plot(np.nan,np.nan),ax[1].set_aspect(1)
    ax[0].legend(loc='center');ax[0],ax[0].set_xlabel('cm'),ax[0].set_ylabel('cm');
    ax[1].legend(loc='center');ax[1],ax[1].set_xlabel('cm'),ax[0].set_ylabel('cm');

px2inch = 1/plt.rcParams['figure.dpi']
plt.rc('legend',fontsize = 16)
```

```python
# @title  Screens H

DATA = [
        #{'size':18.0,'ratio':16/10,'name':'Razer18','color':'k'},
        #{'size':17.0,'ratio':16/10,'name':'LG17' ,'color':'r'},
        {'size':16.0,'ratio':16/10,'name':'Dell16','color':'b'},
        {'size':17.3,'ratio':16/9,'name':'MSI17' ,'color':'m'},
         {'size':15.6,'ratio':16/9,'name':'GELO' ,'color':'g'},
        ]

fig, ax = plt.subplots(1,2,figsize=(1800*px2inch , 1200*px2inch),constrained_layout=True)
for data in DATA:
  plot_rectangle(data,ax)
```
