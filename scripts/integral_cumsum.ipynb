```python
# @title  Screens Ratios

import matplotlib.pyplot as plt
from   matplotlib.patches import Rectangle
import numpy as np

%matplotlib widget


px2inch = 1/plt.rcParams['figure.dpi']
plt.rc('legend',fontsize = 16)
```

```python
g = -9.81

def get_a(t,a=g):           return a*np.ones_like(t)
def get_v(t,a=g,v0=0):      return      v0   + a*t
def get_y(t,a=g,v0=0,y0=0): return y0 + v0*t + a*t**2/2

t  = np.linspace(0,2.038,10000)
dt = np.diff(t)[0]
 
v0 = 10
y0 = 500

a=get_a(t,a=g)
v=get_v(t,a=g,v0=v0)
y=get_y(t,a=g,v0=v0,y0=y0)

vv = np.cumsum(a)*dt + v0
yy = np.cumsum(v)*dt + y0

fig, ax = plt.subplots(1,2,figsize=(800*px2inch , 400*px2inch),constrained_layout=True)

ax[0].plot(t,v,'r')
ax[0].plot(t,vv,'b-.')
ax[0].set_box_aspect(1)
ax[0].set_xlabel('t')
ax[0].set_ylabel('v')


ax[1].plot(t,y,'r')
ax[1].plot(t,yy,'b-.')
ax[1].set_box_aspect(1)
ax[1].set_xlabel('t')
ax[1].set_ylabel('y')
```

```python
def get_seconds(h=0,m=3,s=26,d=0.0): 
    return h*3600 + m*60 + s

dr = [
      [800  ,get_seconds(h=0,m=1, s=41,d=91)],
      [1500 ,get_seconds(h=0,m=3, s=26,d=0 )],
      [3000 ,get_seconds(h=0,m=7, s=17,d=55)],
      [5000 ,get_seconds(h=0,m=12,s=35,d=36)],
      [10000,get_seconds(h=0,m=26,s=11,d=0 )],
      ]

fig, ax = plt.subplots(1,2,figsize=(600*px2inch , 300*px2inch),constrained_layout=True)
k       = np.arange(10)/2
for dri,tri in dr:
    ti = tri + k
    di  = dri*(1-tri/ti)
    dip = 100*(1-tri/ti)
    ax[0].plot(k,di,label=f'{dri}',marker = 'o',ms=5)
    ax[1].plot(k,dip,label=f'{dri}',marker = 'o',ms=5)
    
ax[0].legend()
ax[0].grid(True)
ax[1].legend()
ax[1].grid(True)

plt.show()
```
